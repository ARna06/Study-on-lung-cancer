```python
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import seaborn as sns

# Time-series / stats
import yfinance as yf
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.stattools import adfuller, kpss
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.statespace.sarimax import SARIMAX
import pmdarima as pm                     # auto_arima

# Change-point detection
import ruptures as rpt

# Deep learning
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Reproducibility
np.random.seed(42)
tf.random.set_seed(42)

plt.style.use("seaborn-v0_8-whitegrid")
COLORS = {"blue": "#1f77b4", "orange": "#ff7f0e", "red": "#d62728",
          "green": "#2ca02c", "purple": "#9467bd"}


# ══════════════════════════════════════════════════════════
# SECTION 1 — DATA ACQUISITION & EDA
# ══════════════════════════════════════════════════════════

def fetch_data(tickers: list[str], start: str = "2018-01-01",
               end: str = "2024-12-31") -> pd.DataFrame:
    """Download adjusted close prices from Yahoo Finance."""
    raw = yf.download(tickers, start=start, end=end, auto_adjust=True)["Close"]
    raw.dropna(how="all", inplace=True)
    print(f"✔  Downloaded {len(raw)} rows × {raw.shape[1]} tickers")
    return raw


def compute_returns(prices: pd.DataFrame) -> pd.DataFrame:
    """Log returns and rolling volatility (21-day)."""
    log_ret = np.log(prices / prices.shift(1)).dropna()
    log_ret.columns = [f"{c}_LogRet" for c in log_ret.columns]
    vol = log_ret.rolling(21).std() * np.sqrt(252)
    vol.columns = [c.replace("LogRet", "Vol") for c in vol.columns]
    return pd.concat([log_ret, vol], axis=1).dropna()


def plot_price_and_volume(prices: pd.DataFrame, ticker: str):
    fig, axes = plt.subplots(2, 1, figsize=(14, 6), sharex=True)
    axes[0].plot(prices[ticker], color=COLORS["blue"], lw=1.2)
    axes[0].set_title(f"{ticker} — Adjusted Closing Price", fontsize=13)
    axes[0].set_ylabel("Price (USD)")

    # Rolling 50-day & 200-day moving averages
    axes[0].plot(prices[ticker].rolling(50).mean(),
                 color=COLORS["orange"], lw=1, linestyle="--", label="MA-50")
    axes[0].plot(prices[ticker].rolling(200).mean(),
                 color=COLORS["red"],    lw=1, linestyle="--", label="MA-200")
    axes[0].legend()

    # Volume proxy: absolute log-returns as a bar chart
    log_r = np.log(prices[ticker] / prices[ticker].shift(1)).dropna()
    axes[1].bar(log_r.index, log_r.abs(), color=COLORS["blue"], alpha=0.5)
    axes[1].set_ylabel("|Log Return|")
    axes[1].set_title("Absolute Log Returns (volatility proxy)")

    plt.tight_layout()
    plt.savefig("01_price_overview.png", dpi=150)
    plt.show()
    print("Saved: 01_price_overview.png")


# ══════════════════════════════════════════════════════════
# SECTION 2 — STATIONARITY TESTS & DECOMPOSITION
# ══════════════════════════════════════════════════════════

def adf_kpss_test(series: pd.Series, label: str = "") -> dict:
    """Run ADF and KPSS stationarity tests; return results dict."""
    adf_stat, adf_p, _, _, adf_cv, _ = adfuller(series, autolag="AIC")
    kpss_stat, kpss_p, _, kpss_cv = kpss(series, regression="c", nlags="auto")

    result = {
        "Series": label,
        "ADF Statistic": round(adf_stat, 4),
        "ADF p-value": round(adf_p, 4),
        "ADF Stationary (5%)": adf_p < 0.05,
        "KPSS Statistic": round(kpss_stat, 4),
        "KPSS p-value": round(kpss_p, 4),
        "KPSS Stationary (5%)": kpss_p > 0.05,
    }
    print(pd.Series(result).to_string())
    return result


def decompose_series(series: pd.Series, period: int = 252,
                     model: str = "additive"):
    """Seasonal decomposition — trend / seasonal / residual."""
    result = seasonal_decompose(series.dropna(), model=model, period=period)

    fig, axes = plt.subplots(4, 1, figsize=(14, 10), sharex=True)
    components = [series, result.trend, result.seasonal, result.resid]
    titles     = ["Observed", "Trend", "Seasonal", "Residual"]
    for ax, comp, title in zip(axes, components, titles):
        ax.plot(comp, color=COLORS["blue"], lw=1)
        ax.set_title(title, fontsize=11)
        ax.set_ylabel("Value")
    plt.suptitle(f"Seasonal Decomposition ({model.title()})", fontsize=14, y=1.01)
    plt.tight_layout()
    plt.savefig("02_decomposition.png", dpi=150)
    plt.show()
    print("Saved: 02_decomposition.png")
    return result


def plot_acf_pacf(series: pd.Series, lags: int = 40):
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 4))
    plot_acf(series.dropna(),  lags=lags, ax=ax1)
    plot_pacf(series.dropna(), lags=lags, ax=ax2)
    ax1.set_title("Autocorrelation Function (ACF)")
    ax2.set_title("Partial ACF (PACF)")
    plt.tight_layout()
    plt.savefig("03_acf_pacf.png", dpi=150)
    plt.show()
    print("Saved: 03_acf_pacf.png")


# ══════════════════════════════════════════════════════════
# SECTION 3 — ARIMA / SARIMA FORECASTING
# ══════════════════════════════════════════════════════════

def fit_auto_arima(series: pd.Series, seasonal: bool = True,
                   m: int = 5) -> pm.arima.ARIMA:
    """Use pmdarima auto_arima to select optimal ARIMA/SARIMA order."""
    print("\n── Auto ARIMA search ──")
    model = pm.auto_arima(
        series,
        start_p=1, start_q=1,
        max_p=4,   max_q=4,
        d=None,                     # let ADF decide differencing
        seasonal=seasonal, m=m,     # m=5 for weekly trading days
        information_criterion="aic",
        stepwise=True,
        suppress_warnings=True,
        error_action="ignore",
        trace=True,
    )
    print(f"\nBest order: {model.order}   Seasonal: {model.seasonal_order}")
    print(model.summary())
    return model


def arima_forecast(series: pd.Series, model, n_periods: int = 60,
                   conf: float = 0.95):
    """Generate out-of-sample forecast with confidence interval."""
    fc, ci = model.predict(n_periods=n_periods, return_conf_int=True,
                           alpha=1 - conf)
    idx = pd.date_range(series.index[-1] + pd.Timedelta(days=1),
                        periods=n_periods, freq="B")  # Business days

    fig, ax = plt.subplots(figsize=(14, 5))
    ax.plot(series[-252:], color=COLORS["blue"],   lw=1.2, label="Historical")
    ax.plot(idx, fc,       color=COLORS["orange"], lw=1.5, label="Forecast")
    ax.fill_between(idx, ci[:, 0], ci[:, 1],
                    color=COLORS["orange"], alpha=0.25,
                    label=f"{int(conf*100)}% CI")
    ax.set_title("ARIMA/SARIMA — Out-of-Sample Forecast", fontsize=13)
    ax.legend()
    plt.tight_layout()
    plt.savefig("04_arima_forecast.png", dpi=150)
    plt.show()
    print("Saved: 04_arima_forecast.png")
    return fc, ci
def arima_backtest(series: pd.Series, model,
                   test_size: int = 60) -> pd.DataFrame:
    """Walk-forward validation on a held-out test set using explicit orders."""
    train = series[:-test_size]
    test  = series[-test_size:]

    # Force statsmodels to use the exact orders discovered by your main auto_arima search
    from statsmodels.tsa.statespace.sarimax import SARIMAX

    # Extract structural parameters explicitly
    order = model.order
    seasonal_order = model.seasonal_order

    print(f"\nRefitting explicit SARIMAX{order}x{seasonal_order} on backtest train split...")

    # Fit the explicit model structure
    sarimax_mod = SARIMAX(train, order=order, seasonal_order=seasonal_order,
                          enforce_stationarity=False, enforce_invertibility=False)
    refit = sarimax_mod.fit(disp=False)

    # Forecast out-of-sample over the test window
    forecast_obj = refit.get_forecast(steps=test_size)
    preds = forecast_obj.predicted_mean

    # Grab confidence intervals safely
    ci_df = forecast_obj.conf_int(alpha=0.05)
    ci = ci_df.values

    preds = pd.Series(preds, index=test.index)

    # Double-check for mathematical stability before metric calculation
    if preds.isna().any():
        print("⚠ Warning: Predictions still contain NaNs. Filling forward as fallback.")
        preds = preds.ffill().fillna(train.iloc[-1])

    mae  = mean_absolute_error(test, preds)
    rmse = np.sqrt(mean_squared_error(test, preds))
    mape = np.mean(np.abs((test - preds) / test)) * 100

    print(f"\nARIMA Back-test ({test_size} days):")
    print(f"  MAE  = {mae:.4f}")
    print(f"  RMSE = {rmse:.4f}")
    print(f"  MAPE = {mape:.2f}%")

    fig, ax = plt.subplots(figsize=(14, 5))
    ax.plot(train[-120:], color=COLORS["blue"],   lw=1.2, label="Train")
    ax.plot(test,         color=COLORS["green"],  lw=1.2, label="Actual")
    ax.plot(preds,        color=COLORS["orange"], lw=1.5, linestyle="--",
            label="ARIMA Pred")
    ax.fill_between(test.index, ci[:, 0], ci[:, 1],
                    color=COLORS["orange"], alpha=0.2)
    ax.set_title(f"ARIMA Back-test  |  MAPE {mape:.1f}%", fontsize=13)
    ax.legend()
    plt.tight_layout()
    plt.savefig("05_arima_backtest.png", dpi=150)
    plt.show()
    print("Saved: 05_arima_backtest.png")

    return pd.DataFrame({"Actual": test, "Predicted": preds})


# ══════════════════════════════════════════════════════════
# SECTION 4 — LSTM FORECASTING
# ══════════════════════════════════════════════════════════

def create_sequences(data: np.ndarray, window: int = 60):
    """Convert 1D array into (X, y) supervised pairs."""
    X, y = [], []
    for i in range(window, len(data)):
        X.append(data[i - window:i, 0])
        y.append(data[i, 0])
    return np.array(X)[..., np.newaxis], np.array(y)


def build_lstm(window: int = 60) -> Sequential:
    model = Sequential([
        LSTM(64, return_sequences=True, input_shape=(window, 1)),
        Dropout(0.2),
        LSTM(64, return_sequences=False),
        Dropout(0.2),
        Dense(32, activation="relu"),
        Dense(1),
    ])
    model.compile(optimizer="adam", loss="mse")
    model.summary()
    return model


def train_lstm(series: pd.Series, window: int = 60,
               test_size: int = 60, epochs: int = 50,
               batch_size: int = 32):
    """Scale, build sequences, train LSTM, evaluate."""
    scaler = MinMaxScaler()
    scaled = scaler.fit_transform(series.values.reshape(-1, 1))

    X, y = create_sequences(scaled, window)
    split = len(X) - test_size
    X_train, X_test = X[:split], X[split:]
    y_train, y_test = y[:split], y[split:]

    model = build_lstm(window)
    es = EarlyStopping(monitor="val_loss", patience=10,
                       restore_best_weights=True)

    history = model.fit(
        X_train, y_train,
        validation_split=0.1,
        epochs=epochs,
        batch_size=batch_size,
        callbacks=[es],
        verbose=1,
    )

    # ── training curve
    fig, ax = plt.subplots(figsize=(10, 4))
    ax.plot(history.history["loss"],     label="Train Loss")
    ax.plot(history.history["val_loss"], label="Val Loss")
    ax.set_title("LSTM — Training & Validation Loss")
    ax.set_xlabel("Epoch"); ax.set_ylabel("MSE")
    ax.legend()
    plt.tight_layout()
    plt.savefig("06_lstm_loss.png", dpi=150)
    plt.show()
    print("Saved: 06_lstm_loss.png")

    # ── predictions
    y_pred_s = model.predict(X_test)
    y_pred   = scaler.inverse_transform(y_pred_s).flatten()
    y_true   = scaler.inverse_transform(y_test.reshape(-1, 1)).flatten()

    mae  = mean_absolute_error(y_true, y_pred)
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    r2   = r2_score(y_true, y_pred)
    mape = np.mean(np.abs((y_true - y_pred) / y_true)) * 100

    print(f"\nLSTM Test-set Metrics:")
    print(f"  MAE  = {mae:.4f}")
    print(f"  RMSE = {rmse:.4f}")
    print(f"  R²   = {r2:.4f}")
    print(f"  MAPE = {mape:.2f}%")

    idx = series.index[-test_size:]
    fig, ax = plt.subplots(figsize=(14, 5))
    ax.plot(idx, y_true, color=COLORS["blue"],   lw=1.2, label="Actual")
    ax.plot(idx, y_pred, color=COLORS["red"],    lw=1.5, linestyle="--",
            label="LSTM Predicted")
    ax.set_title(f"LSTM Forecast  |  MAPE {mape:.1f}%  |  R² {r2:.3f}",
                 fontsize=13)
    ax.legend()
    plt.tight_layout()
    plt.savefig("07_lstm_forecast.png", dpi=150)
    plt.show()
    print("Saved: 07_lstm_forecast.png")

    return model, scaler, {
        "MAE": mae, "RMSE": rmse, "R2": r2, "MAPE": mape
    }


# ══════════════════════════════════════════════════════════
# SECTION 5 — CHANGE POINT DETECTION
# ══════════════════════════════════════════════════════════

def detect_change_points(series: pd.Series, n_bkps: int = 5,
                         model_type: str = "rbf"):
    """
    Ruptures library — PELT and Binseg algorithm.
    Correctly maps subplots to avoid matplotlib unpack errors.
    """
    signal = series.values

    # ── PELT (exact, penalized)
    algo_pelt = rpt.Pelt(model=model_type).fit(signal)
    bkps_pelt = algo_pelt.predict(pen=10)

    # ── Binary Segmentation (approximate, fast)
    algo_binseg = rpt.Binseg(model=model_type).fit(signal)
    bkps_binseg = algo_binseg.predict(n_bkps=n_bkps)

    # ── Plot Setup
    fig, axes = plt.subplots(2, 1, figsize=(14, 8), sharex=True)

    algorithms = [
        (bkps_pelt, "PELT (Penalized)", axes[0]),
        (bkps_binseg, f"Binary Segmentation (n={n_bkps})", axes[1])
    ]

    for bkps, title, ax in algorithms:
        # Manually plot the series and breakpoints instead of rpt.display
        ax.plot(series.index, signal, color=COLORS["blue"], lw=1, label="Signal")
        ax.set_title(f"Change Point Detection — {title}", fontsize=12)
        ax.set_ylabel("Price")

        # Annotate break-point dates cleanly
        for bp in bkps[:-1]:
            if bp < len(series):
                date = series.index[bp]
                ax.axvline(x=date, color="crimson", lw=1.5, linestyle="--")
                ax.text(date, ax.get_ylim()[1] * 0.95, # Adjust y-position slightly
                        date.strftime("%b '%y"),
                        fontsize=8, color="crimson", rotation=45, ha='left', va='top')
        ax.legend()

    plt.tight_layout()
    plt.savefig("08_change_points.png", dpi=150)
    plt.show()
    print("Saved: 08_change_points.png")

    # ── Segment-level statistics
    print("\nPELT breakpoints (indices):", bkps_pelt)
    segments = []
    prev = 0
    for bp in bkps_pelt:
        seg = series.iloc[prev:bp]
        segments.append({
            "Start": series.index[prev].strftime("%Y-%m-%d"),
            "End":   series.index[min(bp - 1, len(series) - 1)].strftime("%Y-%m-%d"),
            "Mean":  round(seg.mean(), 2),
            "Std":   round(seg.std(),  2),
            "Trend": "Up" if seg.iloc[-1] > seg.iloc[0] else "Down",
        })
        prev = bp
    seg_df = pd.DataFrame(segments)
    print("\nSegment Statistics:\n", seg_df.to_string(index=False))
    return bkps_pelt, bkps_binseg, seg_df

# ══════════════════════════════════════════════════════════
# SECTION 6 — INTER-ASSET CORRELATION ANALYSIS
# ══════════════════════════════════════════════════════════

def rolling_correlation(prices: pd.DataFrame, window: int = 63):
    """
    63-day (≈ quarterly) rolling Pearson correlation between
    all ticker pairs — reveals evolving co-movement.
    """
    log_ret = np.log(prices / prices.shift(1)).dropna()
    tickers = log_ret.columns.tolist()

    # Static heatmap
    corr = log_ret.corr()
    fig, ax = plt.subplots(figsize=(8, 6))
    mask = np.triu(np.ones_like(corr, dtype=bool))
    sns.heatmap(corr, annot=True, fmt=".2f", cmap="RdYlGn",
                vmin=-1, vmax=1, mask=mask, ax=ax,
                linewidths=0.5, linecolor="white")
    ax.set_title("Full-period Log-Return Correlation", fontsize=13)
    plt.tight_layout()
    plt.savefig("09_correlation_heatmap.png", dpi=150)
    plt.show()
    print("Saved: 09_correlation_heatmap.png")

    # Rolling correlation for each pair
    pairs = [(tickers[i], tickers[j])
             for i in range(len(tickers))
             for j in range(i + 1, len(tickers))]

    fig, ax = plt.subplots(figsize=(14, 5))
    for (t1, t2) in pairs[:6]:           # cap at 6 pairs to keep it readable
        roll = log_ret[t1].rolling(window).corr(log_ret[t2])
        ax.plot(roll, lw=1, label=f"{t1} vs {t2}")

    ax.axhline(0, color="black", lw=0.8, linestyle="--")
    ax.set_title(f"{window}-day Rolling Pearson Correlation", fontsize=13)
    ax.set_ylabel("Correlation")
    ax.legend(fontsize=8)
    plt.tight_layout()
    plt.savefig("10_rolling_correlation.png", dpi=150)
    plt.show()
    print("Saved: 10_rolling_correlation.png")


# ══════════════════════════════════════════════════════════
# SECTION 7 — MODEL COMPARISON DASHBOARD
# ══════════════════════════════════════════════════════════

def model_comparison_table(arima_metrics: dict, lstm_metrics: dict):
    df = pd.DataFrame([arima_metrics, lstm_metrics],
                      index=["ARIMA/SARIMA", "LSTM"])
    print("\n" + "═" * 50)
    print("      MODEL COMPARISON (Test Set)")
    print("═" * 50)
    print(df.round(4).to_string())
    print("═" * 50)

    fig, axes = plt.subplots(1, 3, figsize=(12, 4))
    metrics = ["MAE", "RMSE", "MAPE"]
    for ax, m in zip(axes, metrics):
        vals = [arima_metrics.get(m, np.nan), lstm_metrics.get(m, np.nan)]
        bars = ax.bar(["ARIMA", "LSTM"], vals,
                      color=[COLORS["blue"], COLORS["orange"]])
        ax.bar_label(bars, fmt=".3f", padding=3)
        ax.set_title(m); ax.set_ylabel(m)
    plt.suptitle("Model Comparison — Forecasting Accuracy", fontsize=13)
    plt.tight_layout()
    plt.savefig("11_model_comparison.png", dpi=150)
    plt.show()
    print("Saved: 11_model_comparison.png")


# ══════════════════════════════════════════════════════════
# MAIN PIPELINE
# ══════════════════════════════════════════════════════════

if __name__ == "__main__":

    # ── 1. Data
    TICKERS = ["^NSEI", "^GSPC", "GC=F", "CL=F"]   # NIFTY50, S&P500, Gold, Oil
    prices = fetch_data(TICKERS)

    primary = prices["^NSEI"].dropna()               # main series for modelling

    # ── 2. EDA
    plot_price_and_volume(prices, "^NSEI")
    stats_raw  = adf_kpss_test(primary,                     label="NIFTY50 Price")
    stats_diff = adf_kpss_test(primary.diff().dropna(),     label="NIFTY50 1st Diff")
    log_ret    = np.log(primary / primary.shift(1)).dropna()
    stats_ret  = adf_kpss_test(log_ret,                     label="NIFTY50 Log Return")

    # ── 3. Decomposition
    decompose_series(primary, period=252, model="multiplicative")
    plot_acf_pacf(log_ret, lags=40)

    # ── 4. ARIMA
    arima_model   = fit_auto_arima(primary, seasonal=True, m=5)
    arima_fc, _   = arima_forecast(primary, arima_model, n_periods=60)
    arima_bt      = arima_backtest(primary, arima_model, test_size=60)
    arima_metrics = {
        "MAE":  mean_absolute_error(arima_bt["Actual"], arima_bt["Predicted"]),
        "RMSE": np.sqrt(mean_squared_error(arima_bt["Actual"], arima_bt["Predicted"])),
        "MAPE": np.mean(np.abs((arima_bt["Actual"] - arima_bt["Predicted"])
                                / arima_bt["Actual"]))
    }

    # ── 5. LSTM
    _, _, lstm_metrics = train_lstm(primary, window=60, test_size=60,
                                    epochs=50, batch_size=32)

    # ── 6. Change Point Detection
    bkps_pelt, bkps_binseg, seg_df = detect_change_points(primary, n_bkps=5)

    # ── 7. Correlation
    rolling_correlation(prices, window=63)

    # ── 8. Comparison
    model_comparison_table(arima_metrics, lstm_metrics)

    print("\n✅  All outputs saved. Project complete.")
```
